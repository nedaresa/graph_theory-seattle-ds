```python
# Exercise 1:

# Which state has the highest betweenness centrality?

# HINT: Use the tools inside networkx!


for key, value in nx.betweenness_centrality(borders48).items():
    if value == max(nx.betweenness_centrality(borders48).values()):
        print(key)
```

```python
# Exercise 2:

# Which state has the highest degree centrality?

for key, value in nx.degree_centrality(borders48).items():
    if value == max(nx.degree_centrality(borders48).values()):
        print(key)
```

```python
# Exercise 3:

# Which state has the highest closeness centrality?
for key, value in nx.closeness_centrality(borders48).items():
    if value == max(nx.closeness_centrality(borders48).values()):
        print(key)
```

```python
# Exercise 4:

# Which border has the highest betweenness centrality? 
# And what is the interpretation of this?
#which edge is most used in shortest path

for key, value in nx.edge_betweenness_centrality(borders48).items():
    if value == max(nx.edge_betweenness_centrality(borders48).values()):
        print(key)
```
